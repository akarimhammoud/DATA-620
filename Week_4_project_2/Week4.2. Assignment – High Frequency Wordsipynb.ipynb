### How many total unique words are in the corpus? 

```python
# Select the file Carroll Alice
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')

# Spit words and numerics
alice = [w for w in alice if w.isalpha()]

# Get rid of the stop words
from nltk.corpus import stopwords
alice = [w for w in alice if w not in stopwords.words('english')]
```

```python
# Check the unique words in the dataset
alice_words = set(alice)

#len the words
len(alice_words)
```

### Identify the 200 highest frequency words in this corpus?

```python
# Set up the frequency 
freq_alice = nltk.FreqDist(alice)

# Get the the 200 most common words
alice_200 = pd.DataFrame(freq_alice.most_common(200),columns=['word','count'])

# Get the frequency rate/
alice_200['freq'] = alice_200['count']/float(len(alice))
alice_200.head(10)
```

```python
# lets plot 30 of these common words
freq_alice.plot(30, cumulative = True)
```

### Create a graph that shows the relative frequency of these 200 words?

```python
# plot the 200 most common words
g = sns.barplot(x=alice_200.word, y=alice_200.freq)
```
